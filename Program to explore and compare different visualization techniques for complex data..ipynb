```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import pandas as pd
import numpy as np

df = pd.read_csv('student-por.csv')
print(df.head())

def line_chart(df):
    plt.figure(figsize=(10, 6))
    plt.plot(df['G1'], df['G2'], label='G1 vs G2', color='blue', marker='o', linestyle='-', markersize=6)
    plt.xlabel('G1', fontsize=12)
    plt.ylabel('G2', fontsize=12)
    plt.title('Line Chart: G1 vs G2', fontsize=14)
    plt.legend()
    plt.grid(True)  # Add grid for better readability
    plt.show()


def scatter_plot(df):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df, x='G1', y='G2', hue='sex')
    plt.title('Scatter Plot: G1 vs G2')
    plt.show()

def heatmap(df):
    corr_matrix = df[['G1', 'G2', 'G3']].corr()
    plt.figure(figsize=(8, 6))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
    plt.title('Heatmap: Correlation Matrix')
    plt.show()

def three_d_plot(df):
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(df['G1'], df['G2'], df['G3'], c=df['sex'].astype('category').cat.codes, cmap='viridis')
    ax.set_xlabel('G1')
    ax.set_ylabel('G2')
    ax.set_zlabel('G3')
    ax.set_title('3D Scatter Plot')
    plt.show()

line_chart(df)
scatter_plot(df)
heatmap(df)
three_d_plot(df)

```
