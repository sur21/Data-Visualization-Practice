```python
# Step 1: Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Step 2: Load the dataset from a CSV file
df = pd.read_csv('student-por.csv')  # Replace with the correct file path

# Step 3: Check the first few rows of the dataset to understand its structure
print(df.head())

# Step 4: Create a bar chart using matplotlib
def bar_chart(df):
    plt.figure(figsize=(10, 6))
    # Example: Plot counts of the 'sex' column (categorical data)
    df['sex'].value_counts().plot(kind='bar', color='skyblue')
    plt.title('Bar Chart: Gender Counts')
    plt.xlabel('Gender')
    plt.ylabel('Counts')
    plt.xticks(rotation=45)
    plt.show()

# Step 5: Generate a histogram using seaborn
def histogram(df):
    plt.figure(figsize=(10, 6))
    # Example: Plot the distribution of 'G1' (numerical data)
    sns.histplot(df['G1'], kde=True, color='blue')
    plt.title('Histogram: Distribution of G1')
    plt.xlabel('G1')
    plt.ylabel('Frequency')
    plt.show()

# Step 6: Create a box plot using matplotlib
def box_plot(df):
    plt.figure(figsize=(10, 6))
    # Example: Box plot for 'G1' (numerical data)
    df.boxplot(column='G1', vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
    plt.title('Box Plot: G1 Distribution')
    plt.xlabel('G1')
    plt.show()

# Step 7: Generate a pie chart using matplotlib
def pie_chart(df):
    plt.figure(figsize=(8, 8))
    # Example: Plot proportions of 'sex' (categorical data)
    df['sex'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], startangle=90)
    plt.title('Pie Chart: Gender Proportions')
    plt.ylabel('')  # Hides the 'ylabel' for cleaner presentation
    plt.show()

# Step 8: Call the bar chart function with the dataset
bar_chart(df)

# Step 9: Call the histogram function with the dataset
histogram(df)

# Step 10: Call the box plot function with the dataset
box_plot(df)

# Step 11: Call the pie chart function with the dataset
pie_chart(df)
```
